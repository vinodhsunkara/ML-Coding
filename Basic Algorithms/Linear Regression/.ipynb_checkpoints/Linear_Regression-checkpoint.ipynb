```python
import numpy as np              # numerical computations
import pandas as pd             # data manipulation and analysis
import matplotlib.pyplot as plt # plotting and visualization
import statsmodels.api as sm    # statistical modeling
```

```python
data = pd.read_csv("Advertising.csv")
print(data.head())   # View first few rows
```

```python
print(data.info())   # Check data types and missing values
```

```python
# Summary statistics
print(data.describe())
```

```python
# Include all columns
print(data.describe(include='all'))
```

```python
# Visualize relationships between each ad channel and Sales:
plt.scatter(data["TV"], data["Sales"], alpha=0.5)
# the alpha parameter controls the transparency of the plotted markers.
# alpha = 0: Completely transparent (invisible). alpha = 1: opaque (solid).
plt.xlabel("TV Ad Spend")
plt.ylabel("Sales")
plt.title("TV Ad Spend vs Sales")
plt.show()
```

```python
# Visualize relationships between each ad channel and Sales:
plt.scatter(data["Radio"], data["Sales"], alpha=0.5)
# the alpha parameter controls the transparency of the plotted markers.
# alpha = 0: Completely transparent (invisible). alpha = 1: opaque (solid).
plt.xlabel("Radio Ad Spend")
plt.ylabel("Sales")
plt.title("Radio Ad Spend vs Sales")
plt.show()
```

```python
# Visualize relationships between each ad channel and Sales:
plt.scatter(data["Newspaper"], data["Sales"], alpha=0.5)
# the alpha parameter controls the transparency of the plotted markers.
# alpha = 0: Completely transparent (invisible). alpha = 1: opaque (solid).
plt.xlabel("Newspaper Ad Spend")
plt.ylabel("Sales")
plt.title("Newspaper Ad Spend vs Sales")
plt.show()
```

```python
# Predict Sales using TV, Radio, and Newspaper ad spends.ad channel and Sales
X = data[["TV", "Radio", "Newspaper"]]
```

```python
print(X)
```

```python
Y = data["Sales"]
print(Y)
```

```python
# Fit OLS regression model
model = sm.OLS(Y,X).fit()
```

```python
# Print regression summary
print(model.summary())
```

```python
y_pred = model.predict(X)
y_pred_actual = [[Y, y_pred]]
print(y_pred_actual)
```

```python
# You can also pass new data as we follows with sample ad spend data
new_data = pd.DataFrame({
    "TV": [100, 200],
    "Radio": [20, 40],
    "Newspaper": [10, 20]
})
# Get predictions
new_pred = model.predict(new_data)
print(new_pred)
```

```python
# Plot Regression Line (Example: TV vs Sales)
# TV only regression for visualization
X_tv = sm.add_constant(data["TV"])
model_tv = sm.OLS(Y, X_tv).fit()

y_hat = model_tv.predict(X_tv)

plt.scatter(data["TV"], Y, alpha=0.5, label="Data")
plt.plot(data["TV"], y_hat, color="red", label="Regression Line")
plt.xlabel("TV Ad Spend")
plt.ylabel("Sales")
plt.title("Linear Regression: Sales vs TV Advertising")
plt.legend()
plt.show()
```
