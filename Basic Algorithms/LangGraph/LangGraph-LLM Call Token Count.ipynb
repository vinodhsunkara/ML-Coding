```python
from langchain_openai import ChatOpenAI
from langgraph.graph import StateGraph
from typing import TypedDict
```

```python
# Define a minimal State
class State(TypedDict):
    input: str
    output: str
```

```python
def llm_node(state: State):
    model = ChatOpenAI(model="gpt-4o-mini", api_key="....")
    # get the key fir yourself
    response = model.invoke(state["input"])
    return {"output": response.content}
```

```python
def token_counter_node(state: State):
    tokens = state["output"].split()
    return {"output": len(tokens)}
```

```python
workflow = StateGraph(State)

workflow.add_node("LLM_Model", llm_node)
workflow.add_node("Token_Counter", token_counter_node)

workflow.add_edge("LLM_Model", "Token_Counter")
workflow.set_entry_point("LLM_Model")
workflow.set_finish_point("Token_Counter")

app = workflow.compile()
```

```python
result = app.invoke({"input": "What is an AI agent?"})
print(result)
```
